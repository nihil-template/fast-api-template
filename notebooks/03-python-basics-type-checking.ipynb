```python
age = 25name = "홍길동"print(isinstance(age, int))   # Trueprint(isinstance(age, str))   # Falseprint(isinstance(name, str))  # True# 여러 타입 중 하나인지 확인print(isinstance(age, (int, float)))  # True
```

```python
class Animal:    passclass Dog(Animal):    passdog = Dog()# type() 사용print(type(dog) == Dog)      # Trueprint(type(dog) == Animal)  # False (정확히 일치해야 함)# isinstance() 사용print(isinstance(dog, Dog))    # Trueprint(isinstance(dog, Animal)) # True (상속 관계 고려)
```
