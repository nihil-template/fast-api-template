# Python 기초: 함수 기본

## 개요

함수의 기본 개념을 다룹니다. 함수 정의와 호출, 매개변수 사용, 반환값(`return`)과 여러 값 반환 방법을 학습합니다.

```python
# 함수 정의
def greet():
    print("안녕하세요!")

# 함수 호출
greet()  # 안녕하세요!
```

```python
# 매개변수 하나
def greet(name):
    print(f"안녕하세요, {name}님!")

greet("홍길동")  # 안녕하세요, 홍길동님!

# 매개변수 여러 개
def add(a, b):
    result = a + b
    print(f"{a} + {b} = {result}")

add(3, 5)  # 3 + 5 = 8
```

```python
# 값을 반환하는 함수
def add(a, b):
    return a + b

result = add(3, 5)
print(result)  # 8

# 여러 값 반환 (튜플로 반환)
def get_name_and_age():
    return "홍길동", 25

name, age = get_name_and_age()
print(name, age)  # 홍길동 25
```

```python
# 함수 예제
def calculate_area(width, height):
    """사각형의 넓이를 계산합니다."""
    return width * height

area = calculate_area(5, 3)
print(f"넓이: {area}")  # 넓이: 15
```
