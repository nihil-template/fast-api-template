```python
# math_utils.pydef add(a, b):    return a + bdef multiply(a, b):    return a * b# 이 파일이 직접 실행될 때만 실행됨if __name__ == "__main__":    print("math_utils 모듈이 직접 실행되었습니다.")    print(add(3, 5))    print(multiply(3, 5))
```
