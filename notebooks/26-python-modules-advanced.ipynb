# Python 기초: 모듈 고급

## 개요

모듈의 고급 사용법을 다룹니다. `__name__ == "__main__"` 패턴, 모듈 검색 경로(`sys.path`), 모듈 리로드(`importlib.reload`), `dir()` 함수로 모듈 속성 확인을 학습합니다.

```python
# math_utils.py
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

# 이 파일이 직접 실행될 때만 실행됨
if __name__ == "__main__":
    print("math_utils 모듈이 직접 실행되었습니다.")
    print(add(3, 5))
    print(multiply(3, 5))
```

```python
# main.py
import math_utils

# math_utils의 __main__ 블록은 실행되지 않음
result = math_utils.add(10, 20)
print(result)  # 30
```

```python
import sys

# 모듈 검색 경로 확인
print(sys.path)

# 경로 추가
sys.path.append("/custom/path")
```

```python
import importlib
import math_utils

# 모듈 수정 후
importlib.reload(math_utils)
```

```python
import math_utils

# 모듈의 모든 속성 확인
print(dir(math_utils))

# 특정 속성만 필터링
attributes = [attr for attr in dir(math_utils) if not attr.startswith("_")]
print(attributes)  # ['add', 'multiply', 'PI']
```
