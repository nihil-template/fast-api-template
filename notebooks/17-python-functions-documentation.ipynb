```python
def add(a, b):    """    두 숫자를 더하는 함수        Args:        a (int): 첫 번째 숫자        b (int): 두 번째 숫자        Returns:        int: 두 숫자의 합    """    return a + b# 도움말 확인help(add)print(add.__doc__)
```
