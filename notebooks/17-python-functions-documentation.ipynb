# Python 기초: 함수 문서화

## 개요

함수 문서화 방법을 다룹니다. docstring 작성법, 타입 힌트(type hints) 사용법, `help()` 함수로 문서 확인하는 방법을 학습합니다.

```python
def add(a, b):
    """
    두 숫자를 더하는 함수
    
    Args:
        a (int): 첫 번째 숫자
        b (int): 두 번째 숫자
    
    Returns:
        int: 두 숫자의 합
    """
    return a + b

# 도움말 확인
help(add)
print(add.__doc__)
```

```python
def add(a: int, b: int) -> int:
    """두 정수를 더하는 함수"""
    return a + b

def greet(name: str) -> str:
    """인사말을 반환하는 함수"""
    return f"안녕하세요, {name}님!"

# 타입 힌트는 강제되지 않지만, 코드 가독성과 IDE 지원에 도움
result: int = add(3, 5)
greeting: str = greet("홍길동")
```
