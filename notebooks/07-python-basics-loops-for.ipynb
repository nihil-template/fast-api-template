# Python 기초: for 반복문## 개요Python 기초 학습을 위한 실습 노트북입니다.

```python
# 리스트 순회fruits = ["사과", "바나나", "오렌지"]for fruit in fruits:    print(fruit)
```

```python
# 0부터 4까지 (5는 제외)for i in range(5):    print(i)# 시작값과 끝값 지정for i in range(1, 6):    print(i)# 간격 지정for i in range(0, 10, 2):    print(i)# 역순 (10부터 1까지)for i in range(10, 0, -1):    print(i)
```

```python
# 문자열 순회text = "Python"for char in text:    print(char)
```

```python
# 딕셔너리 순회person = {"name": "홍길동", "age": 25, "city": "서울"}# 키만 순회for key in person:    print(key)# 값만 순회for value in person.values():    print(value)# 키-값 쌍 순회for key, value in person.items():    print(f"{key}: {value}")
```

```python
# 구구단 출력for i in range(2, 10):    print(f"{i}단")    for j in range(1, 10):        print(f"{i} x {j} = {i*j}")    print()
```

```python
# 리스트 합계 계산numbers = [1, 2, 3, 4, 5]total = 0for num in numbers:    total += numprint(f"합계: {total}")  # 합계: 15
```
