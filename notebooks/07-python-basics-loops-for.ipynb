# Python 기초: for 반복문

## 개요

`for` 반복문의 사용법을 다룹니다. 리스트, 문자열, 딕셔너리 순회, `range()` 함수 사용법, 중첩 반복문을 학습합니다.

```python
# 리스트 순회
fruits = ["사과", "바나나", "오렌지"]

for fruit in fruits:
    print(fruit)
```

```python
# 0부터 4까지 (5는 제외)
for i in range(5):
    print(i)

# 시작값과 끝값 지정
for i in range(1, 6):
    print(i)

# 간격 지정
for i in range(0, 10, 2):
    print(i)

# 역순 (10부터 1까지)
for i in range(10, 0, -1):
    print(i)
```

```python
# 문자열 순회
text = "Python"
for char in text:
    print(char)
```

```python
# 딕셔너리 순회
person = {"name": "홍길동", "age": 25, "city": "서울"}

# 키만 순회
for key in person:
    print(key)

# 값만 순회
for value in person.values():
    print(value)

# 키-값 쌍 순회
for key, value in person.items():
    print(f"{key}: {value}")
```

```python
# 구구단 출력
for i in range(2, 10):
    print(f"{i}단")
    for j in range(1, 10):
        print(f"{i} x {j} = {i*j}")
    print()
```

```python
# 리스트 합계 계산
numbers = [1, 2, 3, 4, 5]
total = 0

for num in numbers:
    total += num

print(f"합계: {total}")  # 합계: 15
```
