# Python 기초: 함수 스코프

## 개요

변수의 스코프(유효 범위)를 다룹니다. 전역 변수와 지역 변수의 차이, `global`과 `nonlocal` 키워드 사용법을 학습합니다.

```python
# 전역 변수
global_var = "전역 변수"

def my_function():
    # 지역 변수
    local_var = "지역 변수"
    print(global_var)  # 전역 변수 접근 가능
    print(local_var)

my_function()
# print(local_var)  # 오류! 지역 변수는 함수 밖에서 접근 불가
```

```python
count = 0

def increment():
    global count
    count += 1

increment()
print(count)  # 1
```

```python
def outer():
    x = "외부 함수"
    
    def inner():
        nonlocal x
        x = "내부 함수에서 수정"
        print(x)
    
    inner()
    print(x)

outer()
```
