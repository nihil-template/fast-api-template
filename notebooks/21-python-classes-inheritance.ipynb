# Python 기초: 클래스 상속

## 개요

클래스 상속을 다룹니다. 부모 클래스와 자식 클래스, 메서드 오버라이딩, `super()` 함수 사용, 다중 상속을 학습합니다.

```python
# 부모 클래스
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name}가 소리를 냅니다."

# 자식 클래스
class Dog(Animal):
    def speak(self):
        return f"{self.name}가 멍멍 짖습니다."

class Cat(Animal):
    def speak(self):
        return f"{self.name}가 야옹 웁니다."

dog = Dog("바둑이")
cat = Cat("나비")

print(dog.speak())  # 바둑이가 멍멍 짖습니다.
print(cat.speak())  # 나비가 야옹 웁니다.
```

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "개")  # 부모 클래스의 __init__ 호출
        self.breed = breed
    
    def info(self):
        return f"{self.name}는 {self.species} 종의 {self.breed}입니다."

dog = Dog("바둑이", "골든 리트리버")
print(dog.info())  # 바둑이는 개 종의 골든 리트리버입니다.
```

```python
class Flyable:
    def fly(self):
        return "날 수 있습니다."

class Swimmable:
    def swim(self):
        return "수영할 수 있습니다."

class Duck(Flyable, Swimmable):
    def __init__(self, name):
        self.name = name

duck = Duck("도날드")
print(duck.fly())   # 날 수 있습니다.
print(duck.swim())  # 수영할 수 있습니다.
```
