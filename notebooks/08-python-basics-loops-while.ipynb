# Python 기초: while 반복문

## 개요

`while` 반복문의 사용법을 다룹니다. 조건 기반 반복, 무한 루프와 종료 조건, 사용자 입력과 함께 사용하는 방법을 학습합니다.

```python
# 기본 while 문
count = 0

while count < 5:
    print(count)
    count += 1
```

```python
# 무한 루프 (주의!)
count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break
```

```python
# 사용자 입력과 함께 사용
# 사용자 입력을 받아 반복
while True:
    user_input = input("명령을 입력하세요 (종료: quit): ")
    if user_input == "quit":
        break
    print(f"입력한 명령: {user_input}")
```

```python
# 숫자 맞추기 게임
import random

target = random.randint(1, 100)
attempts = 0

while True:
    guess = int(input("1부터 100 사이의 숫자를 맞춰보세요: "))
    attempts += 1
    
    if guess == target:
        print(f"정답입니다! {attempts}번 만에 맞췄습니다.")
        break
    elif guess < target:
        print("더 큰 수입니다.")
    else:
        print("더 작은 수입니다.")
```

```python
# 짝수만 출력
count = 0
while count < 10:
    if count % 2 == 0:
        print(count)
    count += 1
```
